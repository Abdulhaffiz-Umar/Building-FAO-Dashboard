```python
# Import libraries
import pandas as pd 
import numpy as np 
```

```python
# Import data
df = pd.read_csv("E:/Users/FIZZ/Documents/@Hamoye/Building FAO Dashboard/Data/FoodBalanceSheets_E_Africa/FoodBalanceSheets_E_Africa_NOFLAG.csv")
```

```python
df.head()
```

```python
# shape and data types of the data
print(df.shape)
print(df.dtypes)
```

```python
# select numeric columns
df_numeric = df.select_dtypes(include=[np.number])
numeric_cols = df_numeric.columns.values
print(numeric_cols)
```

```python
# select non numeric columns
df_non_numeric = df.select_dtypes(exclude=[np.number])
non_numeric_cols = df_non_numeric.columns.values
print(non_numeric_cols)
```

```python
# Count the number of null values in the dataset
df.isnull().sum()
```

```python
# Percentage of missing dataset.
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Count number of zeros in all columns of Dataframe
for column_name in df.columns:
    column = df[column_name]
    # Get the count of Zeros in column 
    count = (column == 0).sum()
    print('Count of zeros in column ', column_name, ' is : ', count)
```

```python
#Drop null values
df1 = df.dropna(axis = 0)

df1.to_csv("Cleaned_data.csv")
```

```python
# Remove rows with all 0s in a Dataframe


# Count the number of rows to drop
count_row = 0

# get a copy of the data
df_main = pd.DataFrame(df2)

for i in df2.index:
    if df2["Y2010"][i] == 0.0 and df2["Y2011"][i] == 0.0 and df2["Y2012"][i] == 0.0 and df2["Y2013"][i] == 0.0 and df2["Y2014"][i] == 0.0 and df2["Y2015"][i] == 0.0 and df2["Y2016"][i] == 0.0 and df2["Y2017"][i] == 0.0 and df2["Y2018"][i] == 0.0 and df2["Y2019"][i] == 0.0:
        count_row = count_row + 1
        df_main.drop(df_main.index[i])
        
print(count_row)
```
